# US29

**As a Product Owner, I want to run tests for variable-sized inputs to observe graphically the asymptotic behavior of the execution time of the US13 and US14 procedures (excluding graphic visualization).** 
- **AC01: The graphic referring to the asymptotic behavior of the execution running time tests, as a function of the input size, should be presented in a time unit that allows to distinguish the running times of all tested examples (use any package, for example, Gnuplot).**

## Code

```python
import numpy as np
import matplotlib.pyplot as plt

# --- Input Data ---
# x: Number of stations
# y_us13: Execution time in seconds for version US13
# y_us14: Execution time in seconds for version US14
x = np.array([1150, 450, 350, 250, 550, 1050, 1250, 750, 850, 950, 1450, 150, 650, 50, 1350, 900, 1400, 100, 600, 1300, 1200, 700, 1500, 800, 200, 500, 1000, 1100, 400, 300])
y_us13 = np.array([2.552, 0.153, 0.074, 0.027, 0.274, 1.893, 3.229, 0.698, 1.012, 1.401, 2.689, 0.003, 0.240, 0.0, 2.143, 0.626, 2.409, 0.001, 0.189, 1.925, 1.492, 0.295, 2.989, 0.443, 0.007, 0.113, 0.854, 1.135, 0.058, 0.025])
y_us14 = np.array([0.039, 0.005, 0.001, 0.0, 0.003, 0.014, 0.015, 0.004, 0.006, 0.008, 0.019, 0.0, 0.003, 0.0, 0.016, 0.006, 0.017, 0.0, 0.002, 0.015, 0.013, 0.003, 0.020, 0.005, 0.0, 0.001, 0.008, 0.011, 0.001, 0.0])


# --- Linear Regression Calculations for US13 ---
mean_x = np.mean(x)                   # Mean of x values
mean_y13 = np.mean(y_us13)           # Mean of y values for US13
Sxx = np.sum((x - mean_x) ** 2)      # Sum of squares of x deviations
Sxy13 = np.sum((x - mean_x) * (y_us13 - mean_y13))  # Covariance
beta1_13 = Sxy13 / Sxx               # Slope of regression line
beta0_13 = mean_y13 - beta1_13 * mean_x  # Intercept of regression line
Syy13 = np.sum((y_us13 - mean_y13) ** 2)  # Total variation in y
r13 = Sxy13 / np.sqrt(Sxx * Syy13)  # Pearson correlation coefficient (r)
r2_13 = r13 ** 2                    # Coefficient of determination (r²)

# Print regression parameters for US13
print("US13:")
print("Mean x =", mean_x)
print("Mean y =", mean_y13)
print("Sxx =", Sxx)
print("Sxy =", Sxy13)
print("Beta_1 =", beta1_13)
print("Beta_0 =", beta0_13)
print("Regression Line Equation: y =", beta0_13, "+", beta1_13, "* x")
print("Syy =", Syy13)
print("r =", r13)
print("r² =", r2_13)
print()

# --- Linear Regression Calculations for US14 ---
mean_y14 = np.mean(y_us14)           # Mean of y values for US14
Sxy14 = np.sum((x - mean_x) * (y_us14 - mean_y14))  # Covariance
beta1_14 = Sxy14 / Sxx               # Slope
beta0_14 = mean_y14 - beta1_14 * mean_x  # Intercept
Syy14 = np.sum((y_us14 - mean_y14) ** 2)  # Total variation in y
r14 = Sxy14 / np.sqrt(Sxx * Syy14)  # Pearson correlation coefficient (r)
r2_14 = r14 ** 2                    # Coefficient of determination (r²)

# Print regression parameters for US14
print("US14:")
print("Mean y =", mean_y14)
print("Sxy =", Sxy14)
print("Beta_1 =", beta1_14)
print("Beta_0 =", beta0_14)
print("Regression Line Equation: y =", beta0_14, "+", beta1_14, "* x")
print("Syy =", Syy14)
print("r =", r14)
print("r² =", r2_14)
print()

# --- Prediction Using Regression Line ---
# Predicted y values for both US13 and US14 using their respective regression equations
y_pred13 = beta0_13 + beta1_13 * x
y_pred14 = beta0_14 + beta1_14 * x

# --- Plotting the Results ---
plt.figure(figsize=(12, 6))

# Scatter plot of actual data points and regression lines
plt.scatter(x, y_us13, color='blue', label='US13 - Measured')
plt.plot(x, y_pred13, color='blue', linestyle='--', label=f'US13 - Regression ($R^2$={r2_13:.3f})')

plt.scatter(x, y_us14, color='green', label='US14 - Measured')
plt.plot(x, y_pred14, color='green', linestyle='--', label=f'US14 - Regression ($R^2$={r2_14:.3f})')

# Chart formatting
plt.title("Asymptotic Execution Time – US13 vs US14")
plt.xlabel("Number of Stations")
plt.ylabel("Execution Time (s)") 
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

# US30

**As a Product Owner, I want to obtain the best-fitting curve for the previous sample, and see it graphically.**
- **AC01: The fitted curve should align with the reference function obtained from the theoretical analysis of the procedure’s complexity.**

## Code

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
from scipy.optimize import curve_fit

# Read Data from csv for US13
us13_df = pd.read_csv("US29/US29_csv_results/US13_results.csv")
size_us13 = us13_df["size"].to_numpy()
time_us13 = us13_df["time"].to_numpy()

# Read Data from csv for US14
us14_df = pd.read_csv("US29/US29_csv_results/US14_results.csv")
size_us14 = us14_df["size"].to_numpy()
time_us14 = us14_df["time"].to_numpy()

# Replace zeros with very small values to avoid issues with log
time_us13[time_us13 == 0] = 1e-6
time_us14[time_us14 == 0] = 1e-6

# Corrected mathematical models
def linear(x, a, b):
    return a * x + b

def quadratic(x, a, b, c):
    return a * x**2 + b * x + c

def exponential(x, a, b):
    # Add protection against overflow
    with np.errstate(over='ignore'):
        return a * np.exp(np.clip(b * x, -700, 700))

def logarithmic(x, a, b):
    return a * np.log(x) + b

def n_log_n(x, a, b):
    return a * x * np.log(x) + b

models = {
    "Linear": linear,
    "Quadratic": quadratic,
    "Exponential": exponential,
    "Logarithmic": logarithmic,
    "n log(N)": n_log_n
}

def fit_and_evaluate(x, y):
    results = {}
    for name, model in models.items():
        try:
            if name == "Logarithmic" and np.any(x <= 0):
                results[name] = {"r2": -np.inf, "params": None}
                continue
                
            # Add parameter bounds
            bounds = (-np.inf, np.inf)
            if name == "Exponential":
                bounds = ([0, -np.inf], [np.inf, 0.01])  # Limit exponential growth
                
            params, _ = curve_fit(model, x, y, maxfev=10000, bounds=bounds)
            y_pred = model(x, *params)
            r2 = r2_score(y, y_pred)
            results[name] = {"r2": r2, "params": params, "y_pred": y_pred}
        except Exception as e:
            print(f"Warning: Model fitting failed for {name} - {str(e)}")
            results[name] = {"r2": -np.inf, "params": None}
    return results

# Fit models to US13 and US14
print()
results_13 = fit_and_evaluate(size_us13, time_us13)

print()
results_14 = fit_and_evaluate(size_us14, time_us14)

# Visualization
plt.figure(figsize=(14, 6))
plt.suptitle("Performance Comparison: US13 vs US14", y=1.05)

# US13
plt.subplot(1, 2, 1)
best_13 = max(results_13.items(), key=lambda item: item[1]['r2'])
plt.scatter(size_us13, time_us13, color='blue', alpha=0.6, label='Data')
x_range = np.linspace(min(size_us13), max(size_us13), 100)
plt.plot(x_range, models[best_13[0]](x_range, *best_13[1]['params']), 
         'r-', label=f'{best_13[0]} (R²={best_13[1]["r2"]:.3f})')
plt.title("US13 - Execution Time")
plt.xlabel("Number of Stations")
plt.ylabel("Time (s)")
plt.legend()
plt.grid(True)

# US14
plt.subplot(1, 2, 2)
best_14 = max(results_14.items(), key=lambda item: item[1]['r2'])
plt.scatter(size_us14, time_us14, color='green', alpha=0.6, label='Data')
x_range = np.linspace(min(size_us14), max(size_us14), 100)
plt.plot(x_range, models[best_14[0]](x_range, *best_14[1]['params']), 
         'r-', label=f'{best_14[0]} (R²={best_14[1]["r2"]:.3f})')
plt.title("US14 - Execution Time")
plt.xlabel("Number of Stations")
plt.ylabel("Time (s)")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()  # Removed savefig as requested

# Final results
print("\nResults for US13:")
for name, res in sorted(results_13.items(), key=lambda x: -x[1]['r2']):
    if res['r2'] > -np.inf:
        print(f"{name:10} R²={res['r2']:.4f}")

print("\nResults for US14:")
for name, res in sorted(results_14.items(), key=lambda x: -x[1]['r2']):
    if res['r2'] > -np.inf:
        print(f"{name:10} R²={res['r2']:.4f}")
```

# US31

**As a Player, using the same data file as in the previous Sprint (Sprint 2), I want to perform a statistical analysis to determine which type of cargo contributes most significantly to the revenue of a given station, in a specific year, both defined by the user. The goal for the specified station/year is to identify the cargo (independent variable) that shows the highest correlation with revenue (dependent variable). Based on the selected cargo, a linear regression model should be applied, presenting the error associated with the respective fit. Assuming the amount of cargo increased by 10% in January of the following year, estimate the expected revenue for January with a 95% confidence level.**


## Code

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Read the CSV file using ';' as the separator and ',' as the decimal symbol
df = pd.read_csv('Stations_Data.csv', sep=';', decimal=',')

# Convert text values with commas to decimal numbers, excluding text columns
for col in df.columns:
    if col != 'Station' and df[col].dtype == object:
        df[col] = df[col].str.replace(',', '.', regex=False).str.strip()
        df[col] = pd.to_numeric(df[col], errors='coerce')

# List of available stations
stations = df['Station'].dropna().unique().tolist()

# Display station options
print("\nAvailable Stations:\n")
for i, s in enumerate(stations):
    print(f"{i+1}. {s}")

# User selects a station
station_number = int(input("\nChoose the station number: ")) - 1
chosen_station = stations[station_number]

# List of available years
years = df['Year'].dropna().unique().astype(int).tolist()

# Display year options
print("\nAvailable Years:\n")
for i, y in enumerate(years):
    print(f"{i+1}. {y}")

# User selects a year
year_number = int(input("\nChoose the year number: ")) - 1
chosen_year = years[year_number]

# Filter the dataset for the selected station and year
filtered_data = df[(df['Station'] == chosen_station) & (df['Year'] == chosen_year)]

# Identify relevant cargo columns (excluding unrelated ones)
exclude_cols = ['Station', 'Year', 'Month', 'Arrivals', 'Mail', 'Passengers', 'Revenues', 'Expenses']
cargo_columns = [col for col in filtered_data.columns if col not in exclude_cols]

# Remove rows with missing values in cargo or revenue data
filtered_data = filtered_data.dropna(subset=cargo_columns + ['Revenues'])

# --- CORRELATION ANALYSIS ---
correlations = {}
for cargo in cargo_columns:
    if filtered_data[cargo].std() != 0:
        correlations[cargo] = filtered_data[cargo].corr(filtered_data['Revenues'])

if not correlations:
    print("\nInsufficient data to calculate correlations.")
    exit()

print("\nCorrelation of each cargo type with revenues:")
for cargo, corr in correlations.items():
    print(f"{cargo}: {corr:.4f}")

# Identify cargo type with highest correlation
best_cargo = max(correlations, key=lambda x: abs(correlations[x]))
best_correlation = correlations[best_cargo]

# --- LINEAR REGRESSION CALCULATION ---
x = filtered_data[best_cargo].values
y = filtered_data['Revenues'].values
n = len(x)
mean_x = np.mean(x)
mean_y = np.mean(y)

Sxx = np.sum((x - mean_x) ** 2)
Sxy = np.sum((x - mean_x) * (y - mean_y))
Syy = np.sum((y - mean_y) ** 2)

beta1 = Sxy / Sxx  # Slope
beta0 = mean_y - beta1 * mean_x  # Intercept
y_pred = beta0 + beta1 * x
residuals = y - y_pred
residual_std_error = np.sqrt(np.sum(residuals ** 2) / (n - 2))

# Correlation coefficient and R-squared
r = Sxy / np.sqrt(Sxx * Syy)
r2 = r ** 2

# --- FUTURE FORECAST FOR JANUARY (+10% CARGO) ---
january_data = filtered_data[filtered_data['Month'] == 1]
if not january_data.empty:
    january_value = january_data[best_cargo].values[0]
    future_value = january_value * 1.10
    estimated_revenue = beta0 + beta1 * future_value

    # Confidence interval (95%)
    t_value = stats.t.ppf(0.975, df=n - 2)
    se_fit = residual_std_error * np.sqrt(1 / n + ((future_value - mean_x) ** 2) / Sxx)
    margin_error = t_value * se_fit
    lower_bound = estimated_revenue - margin_error
    upper_bound = estimated_revenue + margin_error
else:
    print("\nNo January data available. Forecast not performed.")
    future_value = estimated_revenue = margin_error = lower_bound = upper_bound = np.nan

# --- RESULTS ---
print("\nUS31 - Simple Linear Regression Statistics")
print(f"Station: {chosen_station}")
print(f"Year: {chosen_year}")
print(f"Type of cargo with the highest correlation with revenue: {best_cargo}")
print(f"Pearson Correlation Coefficient (r): {r:.4f}")
print(f"Coefficient of determination (r²): {r2:.4f}")
print(f"Regression equation: y = {beta0:.4f} + {beta1:.4f}x")
print(f"Residual standard error (s): {residual_std_error:.4f}")

if not january_data.empty:
    print("\nForecast for next January (+10% cargo):")
    print(f"Estimated revenue: €{estimated_revenue:.2f} million")
    print(f"95% Confidence Interval: [€{lower_bound:.2f}, €{upper_bound:.3f}] million")

# --- PLOT ---
plt.figure(figsize=(10, 5))
plt.scatter(x, y, color='blue', label='Observed data')
plt.plot(x, y_pred, color='red', linestyle='--', label=f'Regression ($R^2$={r2:.3f})')
plt.title(f"{chosen_station} - {best_cargo} vs Revenue ({chosen_year})")
plt.xlabel(best_cargo)
plt.ylabel('Revenues (€ million)')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
